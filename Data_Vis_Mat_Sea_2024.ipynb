## Plotting for rest of two products


```python
from matplotlib import pyplot as plt
plt.figure(figsize=(12,4))
plt.plot(df_sales['Quarter'],df_sales['Fridge'])
plt.plot(df_sales['Quarter'],df_sales['Dishwasher'])
plt.plot(df_sales['Quarter'],df_sales['Washing Machine'])
plt.show()
```

## Setting the color for 3 different products

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(12,4))
plt.plot(df_sales['Quarter'],df_sales['Fridge'],color='blue')
plt.plot(df_sales['Quarter'],df_sales['Dishwasher'],color='green')
plt.plot(df_sales['Quarter'],df_sales['Washing Machine'],color='orange')
plt.show()
```

# Giving labels to x-axis,y-axis,title of the graph

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(12,4))
plt.plot(df_sales['Quarter'],df_sales['Fridge'],color='blue')
plt.plot(df_sales['Quarter'],df_sales['Dishwasher'],color='green')
plt.plot(df_sales['Quarter'],df_sales['Washing Machine'],color='orange')
plt.title("Product Sales ")
plt.ylabel("Revenue (mln $)")
plt.xlabel("Financial Quarter")
plt.show()
```

## Labeling to lines which r indicating the products



```python
from matplotlib import pyplot as plt
plt.figure(figsize=(12,4))
plt.plot(df_sales['Quarter'],df_sales['Fridge'],color='blue',label='Fridge')
plt.plot(df_sales['Quarter'],df_sales['Dishwasher'],color='green',label='Dishwasher')
plt.plot(df_sales['Quarter'],df_sales['Washing Machine'],color='orange',label='Washing Machine')
plt.title("Product Sales ")
plt.ylabel("Revenue (mln $)")
plt.xlabel("Financial Quarter")
plt.legend()
plt.show()
```

# Pie chart 

## Getting total sales of each product

```python
total_Sales= df_sales[["Fridge","Dishwasher","Washing Machine"]].sum()
total_Sales
```

```python
total_Sales.index
```

```python
plt.pie(total_Sales,labels=total_Sales.index)
```

```python
plt.pie(total_Sales,labels=total_Sales.index)
plt.show()
```

# Displaying the value in % format using autopct

```python
plt.pie(total_Sales,labels=total_Sales.index,autopct="%1.1f%%")
plt.show()
```

```python
## to get percentage upto 2 fraction

plt.pie(total_Sales,labels=total_Sales.index,autopct="%1.2f%%")
plt.show()
```

# Excluding 1 pie or cutting pie as per value type using explode function

```python
plt.pie(total_Sales,labels=total_Sales.index,autopct="%1.2f%%",explode=(0.1,0,0),shadow= True)
plt.show()
```

## setting angle for that cut piece of pie

```python
plt.pie(total_Sales,labels=total_Sales.index,autopct="%1.2f%%",explode=(0.1,0,0),shadow= True,startangle=140)
plt.show()
```

# Histogram

```python
df_score=pd.read_excel("C:/Users/MANISHA/chapter_3_assets/8_matplotlib_seaborn/histograms.xlsx")
```

```python
df_score
```

```python
plt.hist(df_score['Exam_Score'])
plt.show()
```

```python
import seaborn as sns 
sns.histplot(df_score['Exam_Score'])
plt.show()
```

## to get curve(distribution curve) using kde function [kernel density estimator]

```python
sns.histplot(df_score['Exam_Score'],kde=True)
plt.title("Exam_Scores")
plt.xlabel("Score")
plt.ylabel("Count")
plt.show()
```

## Another way to get histogram using seaborn

```python
sns.histplot(data=df_score,x='Exam_Score',kde=True)
plt.title("Exam_Scores")
plt.xlabel("Score")
plt.ylabel("Count")
plt.show()
```

## Scatter plot

```python
sns.scatterplot(data=df,x="area_square_ft",y="price")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.scatterplot(data=df,x="area_square_ft",y="price")
plt.title("Price Analysis")
plt.xlabel("Area/Square_feet")
plt.show()
```

# Regression Plot

```python
sns.regplot(data=df, x='area_square_ft', y='price', scatter=False, color='red')
plt.xlabel('Area (square feet)')
plt.ylabel('Price')
plt.title('Scatterplot wiath Regression Line')
plt.show()
```
