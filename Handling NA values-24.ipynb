### filing missing values with 0

```python
df.fillna(0,inplace=True)
```

### another method to replace NA values [ffill],[bfill]

## you can set limit for forward and backward fill as per below

```python
df.fillna(method='bfill',limit=2)
```

```python
df.fillna(method='ffill',limit=1)
```
