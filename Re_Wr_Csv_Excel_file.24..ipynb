```python
df_movies=pd.read_excel("C:/Users/MANISHA/Desktop/codebasicExcelNew/movies_db.xlsx","movies")
```

```python
df_movies
```

```python
df_financials=pd.read_excel("C:/Users/MANISHA/Desktop/codebasicExcelNew/movies_db.xlsx","financials")
```

```python
df_mo_act=pd.read_excel("C:/Users/MANISHA/Desktop/codebasicExcelNew/movies_db.xlsx","movie_actor")
df_mo_act.head()
```

```python
df_actors=pd.read_excel("C:/Users/MANISHA/Desktop/codebasicExcelNew/movies_db.xlsx","actors")
df_actors.head()
```

```python
df_lang=pd.read_excel("C:/Users/MANISHA/Desktop/codebasicExcelNew/movies_db.xlsx","languages")
df_lang.head()
```

```python
df_merged_mo_act=pd.merge(df_mo_act,df_actors,on='actor_id')
df_merged_mo_act.head()
```

```python
# Merging language sheet as well

df_merged_language=pd.merge(df_movies,df_lang,on='language_id')
df_merged_language
```

```python
## all worksheets are combining into single file

df_final = pd.merge(df_movies, df_financials, on='movie_id')  # Merge df_movies and df_financials on 'movie_id' column
df_final = pd.merge(df_final, df_merged_mo_act, on='movie_id')  # Merge df_final with df_merged_mo_act on 'movie_id' column
df_final = pd.merge(df_final, df_merged_language, on='movie_id')  # Merge df_final with df_merged_language on 'movie_id' column
```

```python
df_final
```

```python
df_final.shape
```

```python
df_final.to_excel("final.xlsx",sheet_name="final",index=False)
```

```python
## converting currrency from 'dollar' and '$'to "USD"

def standardize_currency(curr):
    if curr == "$$" or curr == "Dollars":
        return "USD"
    return curr

df_financials=pd.read_excel("C:/Users/MANISHA/Desktop/codebasicExcelNew/movies_db.xlsx","financials",converters= {
    'currency': standardize_currency})
df_financials.head()
```

```python
# Merging two sheets into single excel file
df_merged=pd.merge(df_movies,df_financials,on='movie_id')
df_merged.head()
```

```python
df_merged.shape
```

```python
df_merged.to_excel("movies_merged.xlsx",sheet_name="merged",index=False)
```

```python
df_Stocks=pd.DataFrame({
    "tickers": ['GOOGLE','WMT','MSFT'],
     "Price" : [845,65,94],
      "Pe" : [30.37,14.26,30.97],
      "eps": [27.82,4.61,2.12]})
df_Stocks
```

```python
df_weather= pd.DataFrame({
    'Day': ['1/1/2017','1/2/2017','1/3/2017'],
    'Temperature': [32,35,28],
    'Event': ['Rain','Sunny','Snow']})
df_weather
```

```python
# Combining two dataframe into single excel file

with pd.ExcelWriter("stocks_weather.xlsx") as writer:
    df_Stocks.to_excel(writer,sheet_name='stocks')
    df_weather.to_excel(writer,sheet_name='weather')
```
